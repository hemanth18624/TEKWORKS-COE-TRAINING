```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("train.csv")
data.head()
```

```python
#Bar plot of survival rate by passenger class
plt.figure(figsize=(12,6))
sns.barplot(data = data,x = "Pclass",y = "Survived",errorbar = None)
plt.title("Survival Rate of Passengers by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(data = data, x = "Age", y = "Fare", s = 50,hue="Survived",palette = "magma",alpha = 0.9)
plt.title("Fare Price By Age and Gender")
plt.show()
```

```python
plt.figure(figsize=(12,6))
subset_columns = ["Age","Fare","Pclass","Survived"]
sns.pairplot(data = data[subset_columns], hue = "Survived",palette = "plasma")
plt.suptitle("PairwiseRelationships",y = 1.02)
plt.show()
```

```python
plt.figure(figsize=(12,6))
# Select only numeric columns
numeric_columns = data.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
correlation_matrix = numeric_columns.corr()

# Plot the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap="viridis", fmt=".2f")
plt.title("Correlation HeatMap")
plt.show()
```
