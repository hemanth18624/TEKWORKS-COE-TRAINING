```python
import numpy as np
df['CollegeID'] = df['CollegeID'].astype(np.int16)
df['DriveID'] = df['DriveID'].astype(np.int32)
df['CandidateID'] = df['CandidateID'].astype(np.int32)
df['TimeToHire'] = df['TimeToHire'].astype(np.int32)
df['CostPerHire'] = df['CostPerHire'].astype(np.int32)
df['Location'] = df['Location'].astype('category')
```
