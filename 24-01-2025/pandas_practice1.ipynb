```python
data_filled = data.fillna("Unknown")
```

```python
data_filled
```

```python
data_filled.isna().sum()
```

```python
data_filled = data_filled.rename(columns={'Against': 'Opposition'})
data_filled
```

```python
data_filled['Runs'] = data_filled['Runs'].str.replace(r'[^0-9]', '', regex=True)
data_filled
```

```python
data_filled['Runs'] = data_filled['Runs'].astype(int)
```

```python
runs_mean = data_filled['Runs'].mean()
runs_mean
```

```python
no_of_centuries_against_each_team = data_filled.groupby('Opposition')['Runs'].count()
no_of_centuries_against_each_team
```

```python
no_of_centuries_in_each_venue = data_filled.groupby('Venue')['Runs'].count()
no_of_centuries_in_each_venue
```

```python
data_filled.head()
```

```python
number_of_centuries_in_each_position = data_filled.groupby('Position')['Runs'].count()
number_of_centuries_in_each_position
```

```python
number_of_centuries_innings_wise = data_filled.groupby('Innings')['Runs'].count()
number_of_centuries_innings_wise
```

```python
number_of_centuries_based_on_home_or_away = data_filled.groupby('Ground')['Runs'].count()
number_of_centuries_based_on_home_or_away
```

```python
number_of_centuries_in_loosing_or_winning_cause = data_filled.groupby('Result')['Runs'].count()
number_of_centuries_in_loosing_or_winning_cause
```

```python
number_of_centuries = data_filled['No.'].count()
number_of_centuries
```

```python
result_percentage = (number_of_centuries_in_loosing_or_winning_cause / number_of_centuries ) * 100
result_percentage
```
