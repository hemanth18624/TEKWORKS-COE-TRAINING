```python
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import axes3d

plt.style.use('_mpl-gallery')

# Make data
X, Y, Z = axes3d.get_test_data(0.05)

# Plot
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [6,7,8,9,10]

plt.plot(x, y, color="green",linewidth=2,marker='o',linestyle="dashed")
plt.xticks(range(1, 6, 1))
plt.yticks(range(6, 11, 1))



# Adding titles and labels
plt.title("Simple Line Plot")
plt.ylabel("Y-Axis")
plt.xlabel("X-Axis")

plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [1000,3400,2300,6500,4455,6654]
plt.plot(x,y,color="green",linestyle="dashed",marker='o')
plt.xticks(range(1,7,1))
plt.yticks(range(1000,8000,1000))
plt.xlabel("Months")
plt.ylabel("Savings Amount")
plt.title("Savings per Month")
plt.show()
```

```python
categories = ['A','B','C','D','E']
values = [10,43,33,22,67]
plt.bar(categories,values,color=["skyblue","red","yellow","green","orange"],edgecolor="black",width=1)
plt.title("Demonstration of bar chart")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
#Histogram
ages = [20,34,54,23,45,65,32,11,33,21,32]
b = [20,30,40,50]
#bins defines the intervals for grouping the data
plt.hist(ages,bins = b,color="skyblue",edgecolor="black")
plt.grid(False) #Removing the grid boxes
plt.title("Age Distribution of customers")
plt.xlabel("Age Ranges")
plt.ylabel("Frequency")
plt.show()
```

```python
#Pie Chart
language = ["Python","Java","Kotlin","C++"]
percentage = [40,20,24,66]
plt.pie(percentage,labels = language,autopct='%1.1f%%',colors=["skyblue","red","violet","yellow"],startangle=90)
plt.title("Programming Languages Usage")
plt.legend()

plt.show()
```

```python
#Scatter Plot
x = [2,4,3,1,5]
y = [42,46,22,11,55]
plt.scatter(x,y,color="yellow",s= 100, edgecolor="black")
plt.xticks(range(1,6,1))
plt.yticks(range(10,60,10))
plt.title("Scatter Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```
