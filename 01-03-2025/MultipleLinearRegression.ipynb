```python
#Step 1 : Import the required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
#Creating a dataset
data = {

     'Size': [1500, 1600, 1700, 1800, 1900, 2000, 2100, 2200],

     'Bedrooms': [3, 3, 3, 4, 4, 4, 5, 5],

     'Age': [10, 15, 10, 20, 30, 15, 20, 25],

     'Price': [300000, 320000, 340000, 360000, 400000, 370000, 420000, 430000]

 }
df = pd.DataFrame(data)
```

```python
#Define independent variables(features) and dependent variables(target)
X = df[['Size','Bedrooms','Age']] # 3 Independent Variables
Y = df['Price'] # 1 Dependent (Target) Variable
```

```python
#Split the data into training and testing dataset
X_train,X_test,Y_train,Y_test = train_test_split(X,Y)
```

```python
#Create a linear regression model
model = LinearRegression()
```

```python
#Train the Model
model.fit(X_train,Y_train)
```

```python
#Make predictions for the test data set
y_pred = model.predict(X_test)
```

```python
#Coefficients and intercept
print(f"Intercept : {model.intercept_}")
print(f"Coeffieicnt : {model.coef_}")
```

```python
#Evaluate the Model
mse = mean_squared_error(Y_test,y_pred)
print(f"Mean Squared Error : {mse:.2f}")
r_squared_error = r2_score(Y_test,y_pred)
print(f"R Squared Error : {r_squared_error:.2f}")
```

```python
new_house = pd.DataFrame([[1000,2,5]],columns = ['Size','Bedrooms','Age'])



predicted_price = model.predict(new_house)

print(f"Predicted Price for new house: {predicted_price[0]:.2f}")
```
